# Clean Pre2020 Rat Sightings Data

```python
import pandas as pd
```

```python
# List of files and directory
files = [
    "../data/rat_sightings_data/Rat_Sightings_NYC_2010_2012.csv",
    "../data/rat_sightings_data/Rat_Sightings_NYC_2013_2015.csv",
    "../data/rat_sightings_data/Rat_Sightings_NYC_2016_2017.csv",
    "../data/rat_sightings_data/Rat_Sightings_NYC_2018_2019.csv"
]

# Read and concatenate all CSV files
rs = pd.concat([pd.read_csv(file, low_memory=False) for file in files], ignore_index=True)
```

```python
rs
```

```python
# make letters lowercase, replace spaces with underscores, get rid of text after '(' etc

rs.columns = [t.partition('(')[0].strip().lower().replace(' ', '_') for t in rs.columns] #apply to column headers
rs['location_type'] = rs['location_type'].str.strip().str.replace(' ', '_').str.lower()  #apply to location_type column
```

```python
# make the datetime the correct format
rs['created_date'] = pd.to_datetime(rs['created_date']) 
```

```python
rs = rs[['created_date','incident_zip', 'longitude', 'latitude', 'descriptor', 'location_type',
         'unique_key', 'borough']]

rs['date'] = pd.to_datetime(rs['created_date'])
rs = rs.drop(columns=['created_date'])
rs = rs.sort_values(by='date')

rs
```

```python
rs.to_csv("../data/cleaned_rat_sightings_data/cleaned_rat_sightings_pre2020.csv", index=False)
```
