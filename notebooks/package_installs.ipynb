```python
import importlib

# Path to your requirements.txt
requirements_file = "../requirements.txt"  # replace with your path

# Special cases: PyPI package name -> Python import name
special_imports = {
    "scikit-learn": "sklearn",
    "pytorch-lightning": "pytorch_lightning",
}

# Parse the requirements.txt file
expected_versions = {}
with open(requirements_file, "r") as f:
    for line in f:
        line = line.strip()
        if not line or line.startswith("#"):
            continue
        if "==" in line:
            pkg, ver = line.split("==")
            expected_versions[pkg.strip()] = ver.strip()

# Print the header with wider spacing
print(f"{'Package':<25} {'Installed':<20} {'Expected':<15} {'Match?':<8}")
print("-" * 70)

# Check each package is installed and has the correct version
for pkg, expected_version in expected_versions.items():
    import_name = special_imports.get(pkg, pkg.replace("-", "_"))  # handle hyphens automatically
    try:
        module = importlib.import_module(import_name)
        installed_version = getattr(module, "__version__", None)
        if installed_version is None:
            # fallback for unusual cases
            installed_version = getattr(module, "version", None)
        match = "Yes" if installed_version == expected_version else "No"
        print(f"{pkg:<25} {installed_version:<20} {expected_version:<15} {match:<8}")
    except ImportError:
        print(f"{pkg:<25} {'Not installed':<20} {expected_version:<15} No")
```
